```python
def skdj_buy(data, N=9, M=3):
    CLOSE = data['close']
    LOWV = qa.LLV(data['low'], N)
    HIGHV = qa.HHV(data['high'], N)
    RSV = qa.EMA((CLOSE - LOWV) / (HIGHV - LOWV) * 100, M)
    K = qa.EMA(RSV, M)
    D = qa.MA(K, M)
    ind = pd.DataFrame({'IND': pd.concat([K > D, qa.REF(K, 1) < qa.REF(D, 1)], axis=1).all(axis=1) * 1})
    return ind
```

```python
data = candles.data
N = 9
M = 3 
CLOSE = data['close']
LOWV = qa.LLV(data['low'], N)
HIGHV = qa.HHV(data['high'], N)
RSV = qa.EMA((CLOSE - LOWV) / (HIGHV - LOWV) * 100, M)
K = qa.EMA(RSV, M)
D = qa.MA(K, M)

```
